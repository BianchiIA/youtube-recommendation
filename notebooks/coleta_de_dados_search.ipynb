```python
import youtube_dl
import pandas as pd
import numpy as np
```

```python
querys = ["machine+learning", "data+science", "kaggle"]
```

```python
ydl = youtube_dl.YoutubeDL({'ignoreerrors':True})
```

ytsearch:'machine learing'       # Busca normal
"ytsearchdate2:'machine learing  # Busca ordenada por data retorna os 2 primeiros
"ytsearchall:'machine learing  # Busca todos os resultados

```python
r = ydl.extract_info("ytsearchdate2:'machine learing'", download=False)
```

```python
pd.DataFrame(r['entries']).columns
```

```python
pd.DataFrame(r['entries'])
```

```python
resultatos = list()
for query in querys:
    r = ydl.extract_info("ytsearchdate100:'{}'".format(query), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultatos += r['entries']
```

```python
result = [ e for e in resultatos if e is not None]
len(result)
```

```python
df = pd.DataFrame(result)
```

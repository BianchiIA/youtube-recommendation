```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv('../data/raw_data_with_labels.csv', index_col=0)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Selecionando apenas y nao nulo
df = df[df['y'].notnull()]
df.shape
```

```python
df.tail()
```

```python
df.head()
```

```python
df_limpo = pd.DataFrame(index=df.index)
```

# 1. Limpeza dos Dados

```python
claen_date = df['watch-time-text'].str.extract(r"(\d+) de ([a-z]+)\. de (\d+)")
```

```python
claen_date
```

```python
# Tranformando em data do  tipo ddMesAAAA
## Tratando dia
claen_date[0] = claen_date[0].apply(lambda x: '0'+x if len(x) == 1 else x)
```

```python
claen_date[:5]
```

```python
## Tratando mes
mapa_meses = {"jan": "Jan",
              "fev": "Feb",
              "mar": "Mar",
              "abr": "Apr",
              "mai": "May",
              "jun": "Jun",
              "jul": "Jul",
              "ago": "Aug",
              "set": "Sep",
              "out": "Oct",
              "nov": "Nov",
              "dez": "Dec"}
claen_date[1] = claen_date[1].map(mapa_meses)
```

```python
claen_date[:5]
```

```python
claen_date = claen_date.apply(lambda x: " ".join(x), axis=1)
```

```python
claen_date[:5]
```

```python
df_limpo['date'] = pd.to_datetime(claen_date, format='%d %b %Y')
```

```python
df_limpo[:10]
```

# 2. Limpesa de Views

```python
views = df['watch-view-count'].str.extract(r"(\d+\.?\d*)", expand=False).str.replace('.','').fillna(0).astype(int)
df_limpo['views'] = views
```

```python
df_limpo.head().sort_values('date', ascending=False)
```

# 3. Features

```python
features = pd.DataFrame(index=df_limpo.index)
y= df['y'].copy()
```

```python
features['tempo_desde_pub'] = (pd.to_datetime('2019-12-03') - df_limpo['date'])/np.timedelta64(1,'D')
features['views'] = df_limpo['views']
features['views_por_dia'] = features['views'] / features['tempo_desde_pub']
```

```python
features[:10]
```

```python
features.drop('tempo_desde_pub', axis=1, inplace=True)
```

```python
features.head()
```

```python
df_limpo['date'].value_counts().plot(figsize=(20,10))
```

```python
Xtrain, Xval = features[df_limpo['date'] < '2019-04-01'], features[df_limpo['date'] >= '2019-04-01']
ytrain, yval = y[df_limpo['date'] < '2019-04-01'], y[df_limpo['date'] >= '2019-04-01']
```

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier

mdl = DecisionTreeClassifier(random_state=0,
                             max_depth=2,
                             class_weight='balanced')

mdl.fit(Xtrain, ytrain)
```

```python
p = mdl.predict_proba(Xval)[:,1]
p
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score

def eval(y_true, score):
    avg_prec, roc = average_precision_score(y_true,score), roc_auc_score(y_true, score)
    print('Average Precision %s \nROC AUC %s' %(avg_prec, roc))

eval(yval, p)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


plt.figure(figsize=(12,12))
plot_tree(mdl, feature_names=Xval.columns);
```
